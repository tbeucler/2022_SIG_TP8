Pour la derni√®re √©tape, on va faire une boucle qui v√©rifie si le pourcentage de OUI est un nombre au bon format gr√¢ce √† la fonction [`isinstance`](https://docs.python.org/3/library/functions.html#isinstance), c'est √† dire au format [`float`](https://docs.python.org/3/library/stdtypes.html#numeric-types-int-float-complex) correspondant aux [nombres √† virgule flottante](https://fr.wikipedia.org/wiki/Virgule_flottante), et non au format [`string`](https://docs.python.org/3/library/stdtypes.html#text-sequence-type-str) correspondant √† une [cha√Æne de caract√®res](https://fr.wikipedia.org/wiki/Cha%C3%AEne_de_caract%C3%A8res). Par cons√©quent, si le pourcentage de OUI est de type `str`, on efface la ligne.  

Ex√©cute simplement le bloc de code ci-dessous, on t'√©pargne le raisonnement complet üòé

*Remarque: Pour faire la boucle efficacement, on utilise la m√©thode [`iterrows`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html)*

```python
for indice,ligne in donnees.iterrows():
    if isinstance(donnees['Unnamed: 4'][indice],str):
      print('La ligne correspondant √† index=',indice,'a √©t√© effac√©e')
      donnees = donnees.drop(indice,axis=0)

donnees.reset_index(inplace=True) # Remet les indices de ligne dans le bon ordre
donnees.drop("index",axis=1,inplace=True) # enl√®ve la colonne index
```
